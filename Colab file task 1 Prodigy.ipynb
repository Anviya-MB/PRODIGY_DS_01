#Task 1
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution
of ages or genders in a population





```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Data=pd.read_csv('/content/data of task 1.csv')
```

```python
Data.head()
```

```python
Data.tail()
```

```python
#Plotting the data bar chart
plt.bar(Data['Year'],Data['Life Expectancy'])
plt.title('Life Expectancy of people from 2004-2024')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show
```

```python
plt.hist()
```

```python
#Plotting a Histogram
plt.hist(Data['Life Expectancy'])
plt.title('Life Expectancy of people from 2004-2024')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show
```
